## NeuroTKET: An ML-Enhanced Quantum Circuit Compiler Using TKET

This project proposes a novel approach to compiling quantum circuits using machine learning algorithms on Quantinuum's open-source TKET toolchain and API's.

Prior work has shown training an RL agent for these purposes, but we'd like to explore new building a **lightweight learning-based optimizer** that (subject to change):
- Learn circuit simplification patterns
- Adapt to hardware constraints

This project will involve implementing a module that works alongside TKET's native passes. Our outcome is to develop a prototype "neuro-compiler", where we'll measure improvements in
- Gate counts
- Circuit depth
- Expected fidelity

We'll also show a demonstration

### Novelty

1. Avoid heavy pre-training and complex action spaces
2. Leverage TKET's open source optimizations as a baseline
3. Incorporate hardware awareness directly into the compiler
